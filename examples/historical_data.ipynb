# Historical Data

```python
import pandas as pd
from seebuoy import NDBC
```

## Initialize NDBC to pull historical

```python
ndbc = NDBC(timeframe="historical")
```

## Get available data

```python
df_avail = ndbc.available_data(station_id="41002")

print(df_avail.shape)
df_avail.tail(2)
```

## Pull Data

```python
df_data = ndbc.get_data("41002")

print(df_data.shape)
df_data.head(2)
```

```python
df_data.describe()
```

```python
df_data["wave_height"].plot(figsize=(20,5))
```
